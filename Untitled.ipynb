```python
#Product Type, Supplier and Average Manufacturing Cost

defect = df.pivot_table(values = "defect_rates",
                         columns = "product_type",
                         index = "supplier_name",
                         aggfunc = "mean")
print(defect)
defect.plot(kind="bar" , figsize=(8,5))
plt.title("product type vs supplier vs manufacturing cost")
plt.show()
```

```python
# # Q1. Which supplier generates the highest total revenue?

df.groupby("supplier_name")["revenue_generated"] \
    .sum()\
    .sort_values(ascending=False)
```

```python
# Q2. Which product type is the most profitable?
df.groupby("product_type")["revenue_generated"]\
   .sum()\
   .sort_values(ascending=False)
   
```

```python
## Q3. Which supplier has the highest defect rate?

df.groupby("supplier_name")["defect_rates"]\
.mean()\
.sort_values(ascending=False)
```

```python
# Q4. Which shipping carrier has the highest average shipping cost?
df.groupby("shipping_carriers")["shipping_costs"]\
.mean()\
.sort_values(ascending=False)
```

```python
# Q5. Which transportation mode is the most cost-effective?
df.groupby("transportation_modes")["shipping_costs"]\
.mean()\
.sort_values(ascending=False)
```

```python
# Q6. Which products should be restocked first? (Low Stock + High Sales)

df[['sku', 'product_type', 'stock_levels', 'number_of_products_sold']] \
  .sort_values(
      by=['stock_levels', 'number_of_products_sold'],
      ascending=[True, False]
  ) \
  .head(10)
```

```python
# Q7. Which suppliers have high manufacturing costs but low revenue?

df.groupby('supplier_name').agg({
    'manufacturing_costs': 'mean',
    'revenue_generated': 'sum'
}).sort_values(
    by=['manufacturing_costs', 'revenue_generated'],
    ascending=[False, True]
)
```

```python
# Q8. Which routes contribute the highest logistics costs?

df.groupby("routes")["costs"] \
  .sum() \
  .sort_values(ascending=False)
```

```python
df.columns
```

```python
# Q9. Which product type has the best inventory turnover?

df['inventory_turnover'] = (
    df['number_of_products_sold'] / df['stock_levels']
)

df.groupby('product_type')['inventory_turnover'] \
  .mean() \
  .sort_values(ascending=False)
```

```python
## Q10. Which customer demographic generates the highest revenue?
df.groupby("customer_demographics")["revenue_generated"]\
.sum()\
.sort_values(ascending=False)
```
